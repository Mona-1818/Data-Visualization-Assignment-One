```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("haberman.csv",header=None, names = ['Age', 'Operation_Year', 'Positive_Lymph_Nodes', 'Survived'])
```

```python
plt.figure(figsize=(15,10))
plt.plot(data['Age'],data['Positive_Lymph_Nodes'],"ro", markersize=4,label="Positive_Lymph_Nodes")
plt.plot(data['Age'],data['Operation_Year'],"yd", markersize=4,label="Operation_Year")
plt.xlabel("$Age$")
plt.legend(loc="best");
plt.title("scatter plot between Positive_Lymph_Nodes VS Age VS Operation_Year");
```

```python
# Line Chart using seaborn
plt.figure(figsize=(15,10))
sns.lineplot(data=data)
plt.xlabel("$X$")
plt.ylabel("$Y$")
plt.axis([0,350,0,90])
plt.title("LINE CHART");
```

```python
sns.FacetGrid(data,hue = "Survived", height=5).map(plt.scatter, "Age", "Positive_Lymph_Nodes").add_legend()
plt.title("Positive_Lymph_Nodes VS Age scatter plot");
```
